###Create Genre Table

```python
ActorTable = pd.read_pickle('ActorTable_Twitter_clean.pkl')
print ActorTable.shape
MovieTable = pd.read_pickle('MovieTable_Twitter_clean.pkl')
print MovieTable.shape
```

```python
MovieTable.head()
```

```python
GenreTable = pd.DataFrame()
unique_genre = []
for g in MovieTable.Genre:
    for i in g:
       if i not in unique_genre:
            unique_genre.append(i)
print unique_genre
len(unique_genre)
GenreTable["Genre"] = unique_genre
```

```python
Audience_rating = [list([]) for i in xrange(len(unique_genre))]
Critic_rating = [list([]) for i in xrange(len(unique_genre))]
ID = [list([]) for i in xrange(len(unique_genre))]
ReleaseYear = [list([]) for i in xrange(len(unique_genre))]
BoxOffice = [list([]) for i in xrange(len(unique_genre))]
for row_index in xrange(MovieTable.shape[0]):
    i = -1
    for genre in unique_genre:
        i = i + 1
        if genre in MovieTable["Genre"][row_index]:
            ID[i].append(MovieTable["ID"][row_index])
            Audience_rating[i].append(MovieTable["Audience_rating"][row_index])
            Critic_rating[i].append(MovieTable["Critic_rating"][row_index])
            ReleaseYear[i].append(MovieTable["Release"][row_index][0:4])            
            BoxOffice[i].append(MovieTable["BoxOffice"][row_index])        
GenreTable["ID"] = ID
GenreTable["Audience_rating"] = Audience_rating
GenreTable["Critic_rating"] = Critic_rating
GenreTable["ReleaseYear"] = ReleaseYear
GenreTable["BoxOffice"] = BoxOffice
```

```python
Count = []
for i in ID:
    Count.append(len(i))
GenreTable["Count"] = Count
```

```python
import numpy as np
BoxOfficeAvg = []
for i in BoxOffice:
    BoxOfficeAvg.append(np.mean(np.asarray(i)))
GenreTable["BoxOfficeAvg"] = BoxOfficeAvg
```

```python
GenreTable.to_pickle('GenreTable.pkl')
```

### Fixing ActorTable and DirectorTable

```python
ActorTable = pd.read_pickle('ActorTable_Twitter_clean.pkl')
print ActorTable.shape
MovieTable = pd.read_pickle('MovieTable_Twitter_clean.pkl')
print MovieTable.shape
```

```python
Audience_rating = [list([]) for i in xrange(ActorTable.shape[0])]
Critic_rating = [list([]) for i in xrange(ActorTable.shape[0])]
ReleaseYear = [list([]) for i in xrange(ActorTable.shape[0])]
BoxOffice = [list([]) for i in xrange(ActorTable.shape[0])]
Genre = [list([]) for i in xrange(ActorTable.shape[0])]

ActorTable["Audience_rating"] = Audience_rating
ActorTable["Critic_rating"] = Critic_rating
ActorTable["ReleaseYear"] = ReleaseYear
ActorTable["BoxOffice"] = BoxOffice
ActorTable["Genre"] = Genre

for row_index in xrange(ActorTable.shape[0]):
    row_movie = ActorTable["Movie"][row_index]
    for movie_id in row_movie:
        MovieData = MovieTable[MovieTable.ID == movie_id]
        ActorTable["Audience_rating"][row_index].append(MovieData["Audience_rating"].values[0])
        ActorTable["Critic_rating"][row_index].append(MovieData["Critic_rating"].values[0])
        ActorTable["ReleaseYear"][row_index].append(MovieData["Release"].values[0][0:4])
        ActorTable["BoxOffice"][row_index].append(MovieData["BoxOffice"].values[0])
        ActorTable["Genre"][row_index].append(MovieData["Genre"].values[0])
```

```python
ActorTable.to_pickle('ActorTable_Fixed.pkl')
```
