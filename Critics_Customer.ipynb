```python
# get the number of critic ratings for each movie
CriticTable = pd.read_pickle('CriticRaw.pkl')
CriticTable.columns = ['Movie', 'Name', 'Rate', 'Genre', 'BoxOffice', 'Publication']
CriticTable = CriticTable[['Movie', 'Name']].groupby('Movie').count()
CriticTable.reset_index(inplace=True)
CriticTable.columns = ['ID', 'RateCount']
CriticTable.head()
```

```python
len(MovieTable)
```

```python
MovieTable = pd.merge(MovieTable, CriticTable, on='ID')
MovieTable.head()
```

#### Use Bayesian methods to predict audience ratings by critic ratings

```python
#the prior of audience ratings
tau = np.std(MovieTable.Audience_rating, ddof=1)/100
mu = np.mean(MovieTable.Audience_rating)/100

#get the likelihood. As the expert vote is binomial, the std of the likelihood
#can be estimated by the std of binomial dist. 

sigma = np.sqrt(MovieTable.Critic_rating/100*(1-MovieTable.Critic_rating/100)/MovieTable.RateCount)

#estimate the audience rating by the Bayesian method
B = sigma**2/(sigma**2 + tau**2)
audience = B*mu+(1-B)*MovieTable.Critic_rating/100
```

#### Plot the true values against fitted values

```python
plt.scatter(MovieTable.Audience_rating/100, audience, color='black')
```

```python
reg = smf.ols(formula="Audience_rating ~ Critic_rating", data=MovieTable).fit()
plt.scatter(MovieTable.Audience_rating/100, reg.predict()/100)
```

#### When the number of data is small, it is possible that the Bayesian fit is better than linear regression.

```python
#comparison between linear regression and Bayesian predictoin
RSS_linear = sum((reg.resid/100)**2)
RSS_Bayesian = sum((MovieTable.Audience_rating/100-audience)**2)
print 'RSS_linear and RSS_Bayesian'
print RSS_linear, RSS_Bayesian
diff = []
for i in range(2,50):
    select = MovieTable.RateCount<i
    new_audience = audience[select]
    new_reg = smf.ols(formula="Audience_rating ~ Critic_rating", data=MovieTable[select]).fit()
    diff.append(float(sum(abs(new_reg.resid/100)>abs(MovieTable.Audience_rating[select]/100-new_audience)))/sum(select))
plt.scatter(range(2,50), diff)
```
