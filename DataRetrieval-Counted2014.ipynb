###Create Genre Table

```python
import argparse
import json
import pprint
import sys
import urllib
import urllib2
import time
import datetime
import pandas as pd
```

```python
ActorTable = pd.read_pickle('ActorTable_Twitter2014.pkl')
print ActorTable.shape
MovieTable = pd.read_pickle('MovieTable_Twitter2014.pkl')
print MovieTable.shape
```

```python
MovieTable.head()
```

```python
GenreTable = pd.DataFrame()
unique_genre = []
for g in MovieTable.Genre:
    for i in g:
       if i not in unique_genre:
            unique_genre.append(i)
print unique_genre
len(unique_genre)
GenreTable["Genre"] = unique_genre
```

```python
Audience_rating = [list([]) for i in xrange(len(unique_genre))]
Critic_rating = [list([]) for i in xrange(len(unique_genre))]
ID = [list([]) for i in xrange(len(unique_genre))]
ReleaseYear = [list([]) for i in xrange(len(unique_genre))]
BoxOffice = [list([]) for i in xrange(len(unique_genre))]
for row_index in xrange(MovieTable.shape[0]):
    i = -1
    for genre in unique_genre:
        i = i + 1
        if genre in MovieTable["Genre"][row_index]:
            ID[i].append(MovieTable["ID"][row_index])
            Audience_rating[i].append(MovieTable["Audience_rating"][row_index])
            Critic_rating[i].append(MovieTable["Critic_rating"][row_index])
            ReleaseYear[i].append(MovieTable["Release"][row_index][0:4])            
            BoxOffice[i].append(MovieTable["BoxOffice"][row_index])        
GenreTable["ID"] = ID
GenreTable["Audience_rating"] = Audience_rating
GenreTable["Critic_rating"] = Critic_rating
GenreTable["ReleaseYear"] = ReleaseYear
GenreTable["BoxOffice"] = BoxOffice
```

```python
Count = []
for i in ID:
    Count.append(len(i))
GenreTable["Count"] = Count
```

```python
import numpy as np
BoxOfficeAvg = []
for i in BoxOffice:
    BoxOfficeAvg.append(np.mean(np.asarray(i)))
GenreTable["BoxOfficeAvg"] = BoxOfficeAvg
```

```python
GenreTable.to_pickle('GenreTable2014.pkl')
```

### Further Data Wrangling for analysis

```python
import numpy as np
```

```python
ActorTable = pd.read_pickle('ActorTable_Twitter2014.pkl')
print ActorTable.shape
MovieTable = pd.read_pickle('MovieTable_Twitter2014.pkl')
print MovieTable.shape
```

```python
MovieTable.head()
```

```python
Critic_rating_AVG = []
for i in ActorTable["Critic_rating"]:
    Critic_rating_AVG.append(np.mean(np.asarray(i)))
ActorTable["Critic_rating_AVG"] = Critic_rating_AVG

Audience_rating_AVG = []
for i in ActorTable["Audience_rating"]:
    Audience_rating_AVG.append(np.mean(np.asarray(i)))
ActorTable["Audience_rating_AVG"] = Audience_rating_AVG

Boxoffice_AVG = []
for i in ActorTable["Boxoffice"]:
    Boxoffice_AVG.append(np.mean(np.asarray(i)))
ActorTable["Boxoffice_AVG"] = Boxoffice_AVG

Count = []
for i in ActorTable["Movie"]:
    Count.append(len(i))
ActorTable["Count"] = Count
```

```python
ActorTable.to_pickle('ActorTable_Counted2014.pkl')
```

```python
ReleaseYear = []
for i in MovieTable["Release"]:
    ReleaseYear.append(i[0:4])
MovieTable["ReleaseYear"] = ReleaseYear
```
